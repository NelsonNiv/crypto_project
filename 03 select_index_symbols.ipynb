# [assets selection] CRYPTOMC100 Index 

```python
#setting
full_path = '2018-11-01/'
last_date = '2018-06-30'
```

```python
#date base
df = dc.read_pickle(full_path, 'BTC')
df = dc.normalize_df(df, last_date)
df = df[['Market Cap']]
df.rename(columns={'Market Cap':'BTC'}, inplace=True)
#store the coins in array    
array = os.listdir('data/' + full_path)
#prepare lists to store symbols by filter
for symbol in array:
    if symbol != 'BTC':
        df_symbol = dc.read_pickle(full_path, symbol)
        df_symbol = dc.normalize_df(df_symbol, last_date)
        df_symbol = df_symbol[['Market Cap']]
        df_symbol.rename(columns={'Market Cap':symbol}, inplace=True)
        df = df.join(df_symbol)
```

```python
#imprime cantidad de archivos en directorio
print(len(array))
#imprime cantidad de columnas en dataframe
print(len(df.columns))
```

```python
#filtro solo valores de la fecha de cierre estipulada (last date)
df = df[df.index == last_date]
#transponer dataframe
df = df.T.sort_values(by=[last_date], ascending=False)
#guardar dataframe con los 100 activos con mayor capitalización de mercado a la fecha de cierre estipulada
#excluye a estable coins indexados al valor del USD
df = df[(df.index != 'USDT') & (df.index != 'TUSD')][0:100]
```

```python
#data frame d1: 30/06/2018
display(HTML(df.to_html()))
```

```python
#lista 100 assets con mayor capitalización de mercado al 30-06-2018
asset_list = df.index.values
asset_list
```

```python
#imprime largo de lista de assets (comprueba N°)
len(asset_list)
```

```python
#definición de constrains para cada asset
min_large = 180
daily_volume = 40000
coin_presence = 25.0
days_ago = dc.date_days_ago(last_date, 89)
```

```python
#Chequer de estado de data en los 100 crytoassets
checker = dc.apply_filters(full_path, last_date, min_large, days_ago, daily_volume, coin_presence, asset_list)
```

```python
#definir la lista final que compondrá el indice CRYPTOMC100
delist = []
for asset in asset_list:
    df = dc.read_pickle(full_path, asset)
    df = dc.normalize_df(df, last_date)    
    f1 = dc.filter01(df)
    f2 = dc.filter02(df)
    f3 = dc.filter03(df, days_ago)
    f4 = dc.filter04(df,last_date)
    f5 = dc.filter05(df, 'Market Cap', days_ago)
    f6 = dc.filter06(df, min_large, days_ago, daily_volume, coin_presence)
    if 1 in [f1,f2,f3,f4,f5,f6]:
        delist.append(asset)
CRYPTOMC100_assets = [x for x in asset_list if x not in delist]
print(CRYPTOMC100_assets)
```
